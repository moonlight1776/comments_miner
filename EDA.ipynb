```python
text2dict = train['Reviews'].to_dict()
```

```python
# aspect and opinion是否连续
span = {}
for index,row in labels.iterrows():
    if row['id'] not in span:
        span[row['id']] = [] 
    if row['AspectTerms']!='_':
        span[row['id']].append((int(row['A_start']),int(row['A_end'])))
    if row['OpinionTerms']!='_':
        span[row['id']].append((int(row['O_start']),int(row['O_end'])))
for id in span:
    span[id] = sorted(list(set(span[id])),key=lambda x:x[0],reverse=False)
```

```python
# 打印它们
c =0
for index,row in labels.iterrows():
    if row['AspectTerms']!='_' and row['OpinionTerms']!='_':
        a = span[int(row['id'])].index((int(row['A_start']),int(row['A_end'])))
        b = span[int(row['id'])].index((int(row['O_start']),int(row['O_end'])))
        if abs(a-b)>1:
            c+=1
    #print(a,b,span[int(row['id'])],row['id'])
            text = text2dict[int(row['id'])-1]
            print(text)
            
            t = []
            for s in span[int(row['id'])]:
                t.append(text[s[0]:s[1]])
            print(t)
            print(text[int(row['A_start']):int(row['A_end'])],text[int(row['O_start']):int(row['O_end'])])
```
