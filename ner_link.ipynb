```python
import pandas as pd
from sklearn.metrics import f1_score
```

```python
data = pd.read_pickle('result/ner_link_part.pkl')
```

```python
data['p_label'] = data['predict'].apply(lambda x:x>0.002)
```

```python
sum(data['p_label'])
```

```python
data.shape
```

```python
f1_score(data['label'],data['p_label'])
```

```python
data['predict2'] = data.apply(lambda x: x.predict if x.pos1[-1]==x.pos2[-1] else 0,axis=1)
```

```python
for i in range(1000):
    i/=1000
    data['p_label2'] = data['predict2'].apply(lambda x:x>i)
    print(i,f1_score(data['label'],data['p_label2']))
    if i>0.5:break
```
