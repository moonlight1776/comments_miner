```python
test_data = pd.read_pickle('result/ner_link.pkl')
id_list = []
for index, row in test_data.iterrows():
    s = row['text']
    for item in row['result']:
        id_list.append(row['id'])
```

```python
cate['predict_polarity'] = polarity['predict_polarity']
```

```python
cate['id'] = [x+1 for x in id_list]
```

```python
cate
```

```python
def extract_text(s,pos):
    if pos is None:
        return '_'
    else:
        return s[pos[0]:pos[1]+1] 
        
cate['aspect'] = cate.apply(lambda x:extract_text(x.text,x.pos1),axis=1)
cate['opinion'] = cate.apply(lambda x:extract_text(x.text,x.pos2),axis=1)
```

```python
cate.head()
```

```python
complement = []
for i in range(1,2238):
    if i not in cate['id'].unique():
        print(i)
        complement.append([i,'_','_','_','_'])
complement = pd.DataFrame(complement,columns=['id','aspect','opinion','predict_cate','predict_polarity'])
```

```python
submit = cate[['id','aspect','opinion','predict_cate','predict_polarity']]
submit = submit.append(complement)
```

```python
submit = submit.sort_values(by='id')
```
